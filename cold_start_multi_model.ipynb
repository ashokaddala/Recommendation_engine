```python
# converting int64 format into string
items_df['ITEM_ID'] = items_df['ITEM_ID'].astype(str)
```

```python
STOPWORD_LIST = stopwords.words('english') + stopwords.words('portuguese')
#Trains a model whose vectors size is 5000, composed by the main unigrams and bigrams
# found in the corpus, ignoring stopwords
VECTORIZER = TfidfVectorizer(
    input='content',
    encoding='utf-8',
    decode_error='strict',
    strip_accents=None,
    lowercase=True,
    preprocessor=None,
    tokenizer=None,
    analyzer='word',
    stop_words=STOPWORD_LIST,
    token_pattern='(?u)\\b\\w\\w+\\b',
    ngram_range=(1, 2),
    max_df=1.0,
    min_df=1,
    max_features=5000,
    vocabulary=None,
    binary=False,
#     dtype=<class 'numpy.float64'>,
    norm='l2',
    use_idf=True,
    smooth_idf=True,
    sublinear_tf=False,
)
```

```python
ITEM_IDS = items_df['ITEM_ID'].tolist()
#TF-IDF features generation using metadata/factdata
TFIDF_MATRIX = VECTORIZER.fit_transform(items_df['ITEM_TITLE'] + "" + items_df['ITEM_DESCRIPTION'])
```

```python
COSINE_SIM = linear_kernel(TFIDF_MATRIX, TFIDF_MATRIX)
```

```python
def main():
    # save the model to disk
    filename = 'cosine_similarity.pkl'
    pickle.dump(COSINE_SIM, open(filename, 'wb'))
    
if __name__=='__main__':
    main()
```

```python
# load the model from disk
loaded_cold_start_model = pickle.load(open('cosine_similarity.pkl', 'rb'))
```

```python
class coldstartRecommender:
    """it simply recommends to a items based on cosine-similarity
    for the item you want"""
    
    MODEL_NAME = 'Coldstart'
    def __init__(self, coldstart_df, items_daf=None):
        """Initialize an instance variables"""
        self.coldstart_df = coldstart_df
        self.items_daf = items_daf


    def get_model_name(self):
        """returns model name"""
        return self.MODEL_NAME


    def recommend_items(self, item_id, topn=10):
        """returns the similar items based on cosine similarity"""
        item_ids = self.coldstart_df['ITEM_ID']
        item_indices = pd.Series(self.coldstart_df.index, index=self.coldstart_df['ITEM_ID'])
        idx = item_indices[item_id]
        sim_scores = enumerate(loaded_cold_start_model[idx])
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:21]
        item_indices_sim_scores = [i[0] for i in sim_scores]
        similar_items = item_ids.iloc[item_indices_sim_scores]
        #         return similar_items
        recommendations_df = pd.DataFrame(similar_items, columns=['ITEM_ID'])
        ok = recommendations_df.merge(items_df, how='left', left_on='ITEM_ID', right_on='ITEM_ID')[
            ['ITEM_ID']].head(topn)
        return ok
```

```python
coldstart_model = coldstartRecommender(items_df)
```
